```python
from  oceantracker import main

# create parameter as a dictionary
params= {'output_file_base' :'minimal_example',
        'root_output_dir':'output',
        'time_step'  :300, # time step in seconds
        'reader': { 'class_name': 'oceantracker.reader.schism_reader.SCHSIMreaderNCDF',
                'input_dir': 'demo_hindcast',
                'file_mask': 'demoHindcastSchism3D.nc',
                },
        # below releases 10 particles per hour at two given points
        # two 3D release locations        
       'particle_release_groups_list':
                      [{'points': [[1595000, 5483300, -2],[1596000, 5487200, -2] ],
                        'pulse_size': 10, 'release_interval': 3600}
                       ]
                           
        }
```

```python
run_info_file_name, has_errors = main.run(params)

print('name of file useful in plotting',run_info_file_name) 
```

Output is now in  folder output\minimal_example
Fast way to plot is to use the return file name

```python
# import plotting modules needed
from oceantracker.post_processing.read_output_files import load_output_files 
from oceantracker.post_processing.plotting import plot_tracks


# use the run info file to get file name of case infomation json file name as string,
#  which has output file names in it
case_info_file_name = load_output_files.get_case_info_file_from_run_file(run_info_file_name)

# read particle tracks for plotting using case_info_file_name
track_data = load_output_files.load_particle_track_vars(case_info_file_name)

# plot tracks
anim = plot_tracks.plot_tracks(track_data, axis_lims=[1591000, 1601500, 5478500, 5491000],
                            plot_file_name='output\\minimal_example.jpeg')

```

```python
#%matplotlib
from IPython.display import HTML

from matplotlib import pyplot as plt
# need to do amination in seperate window

# animate particles
anim = plot_tracks.animate_particles(track_data, axis_lims=[1591000, 1601500, 5478500, 5491000],
                         title='Minimal example of OceanTracker with 3D point release',
                         movie_file='output\\minimal_example.mp4', show_dry_cells=True)

# below 
HTML(anim.to_html5_video())

```
