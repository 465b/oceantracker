# Comparing grids of the "old" Elbe schism model to the new one

```python
import numpy as np
import netCDF4 as nc
import matplotlib.pyplot as plt
```

```python
hzg_old = '/scratch/local1/hzg3/schout_12.nc'
hzg_new = '/scratch/local1/hzg4/schout_12.nc'

hzg_old = nc.Dataset(hzg_old)
hzg_new = nc.Dataset(hzg_new)
```

```python
x_old = hzg_old.variables['SCHISM_hgrid_node_x'][:]
y_old = hzg_old.variables['SCHISM_hgrid_node_y'][:]
tri_old = hzg_old.variables['SCHISM_hgrid_face_nodes'][:,:3] - 1

x_new = hzg_new.variables['SCHISM_hgrid_node_x'][:]
y_new = hzg_new.variables['SCHISM_hgrid_node_y'][:]
tri_new = hzg_new.variables['SCHISM_hgrid_face_nodes'][:,:3] - 1
```

```python
release_polygon_old = [
    {
        'points': np.array([
					[593043,5919010],
                    [592988,5919053],
					[592924,5918944],
					[592981,5918899]
        ])
    }
]

release_polygon_new = [
    {
        'points': np.array([
            [588957, 5920789],
            [588523, 5920773],
            [588442, 5919759],
            [589038, 5919759],
        ])
    }
]
```

```python
%matplotlib widget

fig, ax = plt.subplots()

# plot old in orange as triplot
ax.triplot(x_old, y_old, tri_old, 'r-', alpha=1, linewidth=2)

# new in black as triplot
ax.triplot(x_new, y_new, tri_new, 'k-', alpha=1, zorder=10)

# plot release polygons as points and arrows in between them
for poly in release_polygon_old:
    ax.scatter(poly['points'][:,0], poly['points'][:,1], c='g', zorder=20)

for poly in release_polygon_new:
    ax.scatter(poly['points'][:,0], poly['points'][:,1], c='g', zorder=20)
    ax.plot(poly['points'][:,0], poly['points'][:,1], 'b-',zorder=20)

# keep aspect ratio 1:1
ax.set_aspect('equal', 'box')

# print clicked coordinates in py array format
def onclick(event):
    print(f'[{event.xdata}, {event.ydata}],')

cid = fig.canvas.mpl_connect('button_press_event', onclick)
```
