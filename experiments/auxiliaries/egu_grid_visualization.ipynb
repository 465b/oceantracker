# drawing grid outline for egu vizualization

```python
from netCDF4 import Dataset
from oceantracker.post_processing.read_output_files.read_ncdf_output_files import read_grid_file
from oceantracker.post_processing.read_output_files.load_output_files import load_grid
from oceantracker.post_processing.plotting.plot_utilities import draw_base_map

import matplotlib.pyplot as plt
```

```python
path_to_dir = '/scratch/local1/output/22_11_01_depth_losses_v07/22_11_01_depth_losses_v07R01_caseInfo.json'
grid = load_grid(path_to_dir)
```

```python
# define axis limits with fixes aspect ratio
window_size = 1e5/4
x_min = 540e3
y_min = 5.915e6
axis_lims = [x_min, x_min+window_size*2, y_min, y_min+window_size]

fig,ax = plt.subplots(figsize=(20,10),dpi=300)
_ = draw_base_map(grid,ax=ax,axis_lims = axis_lims, credit='')

for key, spine in ax.spines.items():
    spine.set_visible(False)

# Alternatively, you can remove specific spines like this:
# ax.spines['top'].set_visible(False)
# ax.spines['right'].set_visible(False)

# Remove ticks and labels
ax.xaxis.set_ticks_position('none')
ax.yaxis.set_ticks_position('none')
ax.xaxis.set_tick_params(labelbottom=False)
ax.yaxis.set_tick_params(labelleft=False)

plt.savefig('harbor.png',dpi=300, transparent=True, bbox_inches='tight', pad_inches=0)
```
