# Calculate growth/death rates

```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.optimize as sio

```

```python
1e-3 == 0.001
```

```python
sign_of_pop_change = +1
fraction_of_particles = 8e-3-5e-3
application_interval = 100#s
```

```python
t = np.arange(0,3600*24*24/24,application_interval)
x = np.zeros(t.shape)
```

```python
x[0] = 1.0
for ii in range(len(x)-1):
    x[ii+1] = x[ii] + x[ii]*fraction_of_particles*sign_of_pop_change
```

```python
plt.plot(t/3600,x)
#plt.xlim()
plt.show()
```

```python
def f(x, A, B):
    return A*np.exp(-B*x)

fit = sio.curve_fit(f, t, x)

plt.plot(t, x)
plt.plot(t, f(t, *fit[0]))
plt.show()
```
