# Replot retention success for EGU 2023

```python
import pickle
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
```

```python
path = '/work/uh0296/u301513/ot_output/22_12_08_retention_v10/retention_success_sa.pkl'
with open(path, 'rb') as f:
    retention_success_sa = pickle.load(f)
    surviving, threshold, split_frac, vert_vel = retention_success_sa[0], retention_success_sa[1], retention_success_sa[2], retention_success_sa[3]
```

```python
fig,ax = plt.subplots(figsize=(10,10),dpi=300)
sns.heatmap(surviving,
                cmap='BrBG',
                center=threshold,
                xticklabels=['{:,.1e}'.format(item) for item in split_frac],
                yticklabels=[round(item,4) for item in vert_vel],square=True
                )
```
