```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
import datetime
sns.set_context('talk')
```

```python
bunthaus = pd.read_csv('dbe_gast_20221114_151817887.csv',delimiter=';',quotechar="'")
seemanshoeft = pd.read_csv('dbe_gast_20221114_152318863.csv',delimiter=';',quotechar="'")

df = pd.concat([bunthaus,seemanshoeft])
df['Stromkilometer'] = [float(item.replace(',','.')) for item in df['Stromkilometer']]
df['Messwert'] = [float(item.replace(',','.')) for item in df['Messwert']]
df['Datum'] = pd.to_datetime(df['Datum'],format="%d.%m.%Y")
```

```python
plt.figure(figsize=(20,10))
plt.grid(True)

dend = datetime.datetime(2012,1,1)
dstart = datetime.datetime(2013,1,1)

plt.scatter(bunthaus['Datum'],bunthaus['Messwert'],label='bunthaus',s=5)
plt.scatter(seemanshoeft['Datum'],seemanshoeft['Messwert'],label='seemanshÃ¶ft',s=5)
plt.legend()
plt.xlim(dstart,dend)
plt.ylabel('FNU')
plt.savefig('long_term_turbidiy.png')
```
